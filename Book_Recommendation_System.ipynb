Popularity Based Recommendation Model

```python
Book_Ratings=ratings.merge(books, on="ISBN")
Book_Ratings.groupby('Book-Title').count()
```

```python
num_rating_df = Book_Ratings.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df = Book_Ratings.groupby('Book-Title')['Book-Rating'].agg(lambda x: x.astype(float).mean()).reset_index()
avg_rating_df.rename(columns = {'Book-Rating' : 'avg_rating'}, inplace = True)
avg_rating_df
```

```python
popular_book_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_book_df
```

```python
popular_book_df = popular_book_df[popular_book_df['num_ratings']>=250].sort_values('avg_rating',ascending=False).head()
popular_book_df = popular_book_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_rating']]
popular_book_df['Image-URL-M'][0]
```

Collaborative Filtering based Recommendation System

```python
x = Book_Ratings.groupby('User-ID').count()['Book-Rating'] > 200
active_users = x[x].index

filtered_rating = Book_Ratings[Book_Ratings['User-ID'].isin(active_users)]
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_book_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_book_ratings.shape
```

```python
final_book_ratings.drop_duplicates(['User-ID','Book-Title'],inplace=True)
```

```python
final_book_ratings.shape
```

```python
pt = final_book_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt.fillna(0,inplace=True)
pt
```

```python
from sklearn.neighbors import NearestNeighbors

def recommend2(book_name, pt, books_data, n_neighbors=4):
    model = NearestNeighbors(metric='cosine', algorithm='brute')
    model.fit(pt)

    query_index = pt.index.get_loc(book_name)
    distances, indices = model.kneighbors([pt.iloc[query_index]], n_neighbors=n_neighbors+1)

    similar_indices = indices.flatten()[1:]  # Exclude the first item (itself)
    similarity_scores = distances.flatten()[1:]  # Exclude the first item (itself)

    recommended_books = []
    for idx, sim_score in zip(similar_indices, similarity_scores):
        book_title = pt.index[idx]
        temp_df = books_data[books_data['Book-Title'] == book_title].drop_duplicates('Book-Title')
        if len(temp_df) > 0:  # Ensure there's matching data
            item = [
                temp_df['Book-Title'].values[0],
                temp_df['Book-Author'].values[0],
                temp_df['Image-URL-M'].values[0]
            ]
            recommended_books.append(item)

    return recommended_books
```
